# Now, we're changing the name of column Marital_Status to indian name Shadi

```python
# changing the column name Marital_Status to Shadi
diwali_sales_data.rename(columns={'Marital_Status':'Shadi'}, inplace = True)
```

```python
# as we see, we changed the name of column
diwali_sales_data.columns
```

## Here we also changing the F and M to Female and Male.

```python
# here we changed F to Female in Gender column
def remove_f(a):
    a = a.replace('F', 'Female')
    return a
 
diwali_sales_data['Gender'] = diwali_sales_data['Gender'].apply(remove_f)

# here we changed M to Male in Gender column
def remove_m(b):
    b = b.replace('M', 'Male')
    return b
 
diwali_sales_data['Gender'] = diwali_sales_data['Gender'].apply(remove_m)
```

```python
# here we changed 0 to Married  in Shadi column
def remove_01(c):
    if c == 0:
        return 'Married'
    else:
        return 'Unmarried'

diwali_sales_data['Shadi'] = diwali_sales_data['Shadi'].apply(remove_01)
```

```python
diwali_sales_data.head()
```

Perfect, we fixed the gender short form and marital status problem. 

# Exploratory Data Analysis

## Firstly, we will check how many customers are male or female so we can figure out who buys the most.

```python
# plotting a bar chart for Gender and the count
gen_bar = sns.countplot(x= 'Gender', data= diwali_sales_data, palette=['#2F3C7E', '#FBEAEB'])

for i in gen_bar.containers:
  gen_bar.bar_label(i)
```

```python
# plotting a bar chart for gender vs the amount
buy_most = diwali_sales_data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = buy_most, palette=['#FCEDDA', '#EE4E34'])
```

*Looking at the above graphs, we can tell that most of the people buying items are females. Additionally, the females seem to be spending more money than the males.*

## Finding the ages of customers based on female and male category and find out who bought items during Diwali sales.

```python
find_age = sns.countplot(data = diwali_sales_data, x = 'Age Group', hue = 'Gender', palette= ['#CCF381','#4831D4'])

for i in find_age.containers:
    find_age.bar_label(i)
```

```python
# finding the spent amount vs age group
spent_age = diwali_sales_data.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group', y= 'Amount' ,data = spent_age, palette= ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#800080', '#FFA500', '#FFC0CB'])
```

*The graphs show that most of the people who bought items during Diwali sales are females between the ages of 26 to 35 years old.*

## Finding the most buyers belongs to which state ? so we can find out most of the orders from which state ?

```python
# buyers from most 10 state
buyers_state = diwali_sales_data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = buyers_state, x = 'State',y= 'Orders', palette= ['#000080', '#008080', '#FF7F50', '#87CEEB', '#D3D3D3', '#808000', '#FFFF99', '#8B0000', '#E6E6FA', '#008B8B'])
```

```python
# finding the total amount/sales from most 10 states
amount_state = diwali_sales_data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = amount_state, x = 'State',y= 'Amount', palette= ['#DC143C', '#006400', '#FF1493', '#1E90FF', '#F08080', '#7B68EE', '#6B8E23', '#A0522D', '#9ACD32', '#BDB76B'])
```

*We can see from the graphs that the highest number of orders and the highest total sales or amount were from Uttar Pradesh, followed by Maharashtra and Karnataka.*


## Finding the marital status and check to find out most buyer based on gender

```python
c = sns.countplot(data = diwali_sales_data, x = 'Shadi', palette= ['#5F4B8BFF', '#E69A8DFF'])

sns.set(rc={'figure.figsize':(7,5)})
for i in c.containers:
    c.bar_label(i)
```

```python
buyers_state = diwali_sales_data.groupby(['Shadi', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = buyers_state, x = 'Shadi',y= 'Amount', hue='Gender', palette= ['#00203FFF', '#ADEFD1FF'])
```

*Based on the graphs, we can observe that the majority of the buyers are married women, and they tend to have a high purchasing power.*

## Finding the Occupation of Customers so we can figure out in which industry most buyers work?

```python
# finding the occupation of the customers
sns.set(rc={'figure.figsize':(20,5)})
occupation = sns.countplot(data = diwali_sales_data, x = 'Occupation')

for i in occupation.containers:
    occupation.bar_label(i)
```

```python
# # finding the buyers industry
buyers_industry = diwali_sales_data.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = buyers_industry, x = 'Occupation',y= 'Amount')
```

*From the graphs, we can see that most of the buyers work in the IT, Healthcare, and Aviation sectors.*

## Finding the Product Category so can figure out most of the sold products from which category.

```python
# finding all categories from products.
product_categories = sns.countplot(data = diwali_sales_data, x = 'Product_Category')
sns.set(rc={'figure.figsize':(30,5)})

for i in product_categories.containers:
    product_categories.bar_label(i)
```

```python
# finding most sold product from which categories 
sold_product_categories = diwali_sales_data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data = sold_product_categories, x = 'Product_Category',y= 'Amount')
```

*Based on the graphs, we can see that the majority of the sold products belong to the Food, Clothing, and Electronics categories.*

## Finding the top 10 product that sold.

```python
# top 10 most sold products (same thing as above)
top_10 = plt.subplots(figsize=(12,7))
diwali_sales_data.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
